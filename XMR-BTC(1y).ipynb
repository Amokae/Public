```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import seaborn

%matplotlib inline
```

```python
XMR  = pd.read_csv('XMR.csv');  XMR['pct'] = XMR['Adj Close'].pct_change()
BTC4  = pd.read_csv('BTC4.csv');  BTC4['pct'] = BTC4['Adj Close'].pct_change()

xmr  = XMR['pct'].to_numpy(); xmr = np.delete(xmr, 0)
btc4  = BTC4['pct'].to_numpy(); btc4 = np.delete(btc4, 0)
```

```python
plt.plot(xmr, btc4, '.')
plt.grid(True)
plt.xlabel('SPY % Change')
plt.ylabel('DELL % change')

slope, intercept, r_value, p_value, std_err =  linregress(xmr, btc4)

x = np.linspace(-.1, .1)
plt.plot(x, slope * x + intercept, 'k')

print('beta = ', slope)
print('Corr = ', r_value)
```
